```python
def plotter(lru, fifo, rands, pagesize, file):
    plt.plot(xss,lru,label='lru')
    plt.plot(xss,fifo,label='fifo')
    plt.plot(xss,rands,label='random')
    plt.title("Input: "+file+"; Page size: "+str(pagesize))
    plt.xlabel("Number of frames")
    plt.ylabel("Number of page faults")
    plt.legend()
    plt.savefig(file+str(pagesize)+".png")

def prepare_array(array):
    split = array.split(' ')
    return [int(i) for i in split]
```

```python
plotter(prepare_array(f256lru), prepare_array(f256fifo), prepare_array(f256rand), 256, "focus.trace")
```

```python
plotter(prepare_array(f1024lru), prepare_array(f1024fifo), prepare_array(f1024rand), 1024, "focus.trace")
```

```python
plotter(prepare_array(s256lru), prepare_array(s256fifo), prepare_array(s256rand), 256, "sort.trace")
```

```python
plotter(prepare_array(s1024lru), prepare_array(s1024fifo), prepare_array(s1024rand), 1024, "sort.trace")
```

```python
plotter(prepare_array(sc256lru), prepare_array(sc256fifo), prepare_array(sc256rand), 256, "scan.trace")
```

```python
plotter(prepare_array(sc1024lru), prepare_array(sc1024fifo), prepare_array(sc1024rand), 1024, "scan.trace")
```
